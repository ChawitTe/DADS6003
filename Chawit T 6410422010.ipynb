```python
# Import the Libraries
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
test_df = pd.read_csv("../input/train-test/test.csv")
train_df = pd.read_csv("../input/train-test/train.csv")
```

```python
train_df.info()
```

```python
# ตรวจสอบค่า Null แต่ละ column
total = train_df.isnull().sum().sort_values(ascending=False)
percent_1 = train_df.isnull().sum()/train_df.isnull().count()*100
percent_2 = (round(percent_1, 1)).sort_values(ascending=False)
missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
missing_data.head(20)
```

```python
# Drop ทุก rows ที่มีค่า NA
train_df =train_df.dropna()
```

```python
# หาความสัมพันธ์เพื่อเลือก features
corrmat = train_df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
sns.heatmap(train_df[top_corr_features].corr(),annot=True,cmap="YlGnBu")
```

```python
# Drop column Y 
X_train = train_df.drop("y", axis=1)
Y_train = train_df["y"]
X_test  = test_df

print(X_train.shape, Y_train.shape, X_test.shape)
```

```python
# เลือก features
X_train = X_train[['x2', 'x3', 'x9', 'x15', 'x14', 'x7', 'x6', 'x10']]
X_test  = X_test[['x2', 'x3', 'x9', 'x15', 'x14', 'x7', 'x6', 'x10']]
print(X_train.shape, Y_train.shape, X_test.shape)
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier()
decision_tree = decision_tree.fit(X_train, Y_train)  
Y_pred = decision_tree.predict(X_test)  
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
```

```python
# Random Forest
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)

Y_prediction = random_forest.predict(X_test)

random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
```

```python
# Neural Network
nn = MLPClassifier()
nn.fit(X_train, Y_train)  
Y_pred = nn.predict(X_test)  
acc_nn = round(nn.score(X_train, Y_train) * 100, 2)
```

```python
# Check Model score
results = pd.DataFrame({
    'Model': ['Decision Tree', 'Random Forest', 'Neural Network'],
    'Score': [acc_decision_tree, acc_random_forest, acc_nn]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df
```

```python
# send out_put into csv.
random_forest = RandomForestClassifier(n_estimators=1000)
random_forest = random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
pred_random_forest = pd.DataFrame({
    "Expected": Y_pred
})
pred_random_forest.insert(0, 'Id', range(1, 1 + len(pred_random_forest)))
pred_random_forest.to_csv('submission_pred_rf.csv', index=False)
```
